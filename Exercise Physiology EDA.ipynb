# HISTOGRAM

```python
sns.histplot(data=data,x="Age",kde=True)
plt.title("histogram")
```

The histogram of health data indicates a positively skewed distribution among age groups ranging from 20 to 80, with a focus on exercise. The 20 to 30 age group shows a high count, suggesting a health-conscious and actively exercising population. However, it's crucial not to overlook older individuals, as they also participate in exercise routines, warranting consideration in health analyses.

```python
sns.histplot(data=data,x="Age",kde=True,bins=7)
```

```python
sns.histplot(data=data,x="Height",kde=True)
plt.title("Histogram")
```

The histogram of health data reveals that the heights of both males and females follow approximately normal distributions. This suggests that the distribution of height values for both genders is centered around a mean, with relatively equal representation on either side of the mean.

```python
sns.histplot(data=data,x="Heart_Rate",color="r",kde=True,binwidth=10)
plt.title("Histogram")
```

The histogram analysis of health data reveals that the distribution of heart rates for both males and females is approximately normal, indicating a balanced and typical pattern in the dataset. This suggests that heart rate values for both genders are spread out in a way that resembles a standard or Gaussian distribution.

```python
sns.histplot(data=data,x="Body_Temp",kde=True)
plt.title("Histogram")
```

The histogram analysis of health data illustrates that the Body Temperature for both males and females exhibits an approximately negative distribution. Moreover, after engaging in exercises, there is a noticeable increase, with values ranging from 800 to 1000, indicating a rise in body temperature post-exercise for both genders.

```python
sns.histplot(data=data, x='Body_Temp', hue="Gender", kde=True);
plt.title("Histogram")
```
